```python
import os
import openai
import config as cfg
from langchain.chat_models import ChatOpenAI
import pandas as pd

# ChatGPT API Key
os.environ['OPENAI_API_KEY'] = cfg.gpt_config["api_key"]
df = pd.read_csv('본회의처리안건_법률안.csv')
```

# langchain 모델 정의

```python
llm = ChatOpenAI(temperature=0,
                 model='gpt-3.5-turbo-1106',
                 )
```

# Labeling Guideline

```python
label_template = """라벨링 가이드라인에 따라 단계적으로 주어진 법률안에 대한 법률 카테고리를 분류하세요.

       라벨은 인권 환경 노동 경제 사회 교통으로 6개 입니다.
       
       * 인권: 민법, 형법, 국가 인권위원회 법, 성폭력 범죄의 처벌 등에 관한 특례법, 인권 기본법, 기타 등등에 대한 사람이 인권에 관련된 법안
       
       * 경제: 공정거래법, 소비자보호법, 금융실명거래 및 비밀보장에 관한 법률, 자본시장과 금융투자업에 관한 법률 등 경제적인 거래에 대해 정의하는 법률

       * 환경: 환경기본법, 자연환경보존법, 수질환경보전법, 대기환경보존법, 폐기물관리법, 화학물질관리법, 소음진동관리법 등등에 대한 환경을 파괴하거나 해치는 것을 방지하기 위한 법안
       
       * 노동: 노동법, 근로기준법, 노동조합 및 노동관계조정법, 산업재해보상보험법 등에 관련한 법률 카테고리로, 인권과 비슷한 성질을 가질 수 있지만, 어떤 직업이나 노동자들과 관련된 법안
              
       * 사회: 청소년 보호법, 아동 복지법, 노동 복지법, 여성발전 기본법, 장애인 복지법, 국민 기초생활 보장법 등에 관련한 법률 카테고리로, 인권과 비슷한 성질을 가질 수 있지만 사회적 약자들을 보호하기 위한 법률
       
       * 교통: 도로교통법, 철도안전법, 항공법, 선박안전법, 자동차관리법, 자동차보험법 등 교통에 관련한 법률

이 6개의 라벨외의 다른 라벨은 없습니다.

"""
```

# GPT를 활용한 데이터 Labeling

```python
df['카테고리'] = None

for i in range(len(df)):
    law_name = df['의안명'].iloc[i]
    question = f"너는 법률 안건에 라벨링을 해주는 데이터 분석가야. 라벨링 가이드라인:{label_template}에 따라 법률안:'{law_name}'에 라벨링을 해주고 답변은 라벨 두 글자만 해줘."
    response = llm.predict(question)
    df.loc[i, '카테고리'] = response
    if i % 10 == 0:
        print(f"의안명:{law_name}, 라벨:{response}")
```

```python
df.to_csv('본회의처리안건_법률안.csv', index=False)
```
