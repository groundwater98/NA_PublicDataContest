```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

file = '본회의처리안건_법률안.xls'
df = pd.read_excel(file)

df.tail()
```

```python
print('전체 데이터 건수:', len(df))
```

```python
print('컬럼별 결측치 개수')
df.isnull().sum().to_frame('nan_count')
```

```python
# 결측치 비율 파악
pd.DataFrame(data=df.isnull().sum()/len(df), columns=['nan_ratio'])
```

```python
df = df.drop(columns=['의안활동구분'])
df = df.drop(columns=['의안번호'])
df = df.drop(columns=['대수'])
df = df.drop(columns=['제안자'])
df = df.drop(columns=['위원회심사_회부일'])
df = df.drop(columns=['법사위체계자구심사_회부일'])
df = df.drop(columns=['법사위체계자구심사_의결일'])
df = df.drop(columns=['의안ID'])
df = df.drop(columns=['소관위원회ID'])
```

```python
df.to_csv('본회의처리안건_법률안.csv', index=False)
```
