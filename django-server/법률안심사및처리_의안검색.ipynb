```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

file = '데이터_법률안 심사 및 처리(의안검색).xlsx'
df = pd.read_excel(file)

df.tail()
```

# 결측치 확인

```python
print('전체 데이터 건수:', len(df))
```

```python
print('컬럼별 결측치 개수')
df.isnull().sum().to_frame('nan_count')
```

```python
# 결측치 비율 파악
pd.DataFrame(data=df.isnull().sum()/len(df), columns=['nan_ratio'])
```

# 제안된 안건은 최신 의안건에 대해서만 필요하므로 2023년 이전 내용들은 삭제

```python
df['제안일'] = pd.to_datetime(df['제안일'], errors='coerce')
# 기준 날짜 설정
cutoff_date = pd.Timestamp('2023-09-01')

# 기준 날짜 이전의 데이터 필터링
df = df[df['제안일'] >= cutoff_date]

# NaN이나 NaT 값을 가진 행 삭제
df = df.dropna(subset=['제안일'])
```

```python
# 결측치 비율 파악
pd.DataFrame(data=df.isnull().sum()/len(df), columns=['nan_ratio'])
```

# 안건들을 최신 날짜 기준으로 정렬

```python
df = df.sort_values(by="제안일", ascending=False)
```

# 데이터 처리를 위한 필요없는 Column 삭제

```python
df = df.drop(columns=['대'])
df = df.drop(columns=['제안자'])
df = df.drop(columns=['의결일'])
df = df.drop(columns=['본회의심의결과'])
df = df.drop(columns=['소관위회부일'])
df = df.drop(columns=['소관위상정일'])
df = df.drop(columns=['소관위처리일'])
df = df.drop(columns=['소관위처리결과'])
df = df.drop(columns=['법사위회부일'])
df = df.drop(columns=['법사위상정일'])
df = df.drop(columns=['법사위처리일'])
df = df.drop(columns=['법사위처리결과'])
df = df.drop(columns=['제안자구분'])
```

```python
df.to_csv('법률안심사및처리_의안검색.csv', index=False)
```
